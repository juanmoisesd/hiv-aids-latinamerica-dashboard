````python
!pip install pandas matplotlib seaborn numpy -q
import pandas as pd,matplotlib.pyplot as plt,seaborn as sns,numpy as np
plt.style.use("seaborn-v0_8-whitegrid")
print("✅ Ready")
````

## Load data

````python
# Full data: https://juanmoisesd.github.io/hiv-aids-latinamerica-dashboard/
np.random.seed(42)
cc=["Brazil","Mexico","Argentina","Colombia","Chile","Peru","Venezuela","Ecuador"]
yy=list(range(2015,2026))
rows=[{"country":c,"year":y,"value":round(np.random.uniform(8,25)+i*0.3+np.random.normal(0,0.5),2),"n":int(np.random.uniform(500,5000))} for c in cc for i,y in enumerate(yy)]
df=pd.DataFrame(rows)
df.head(8)
````

## Visualizations

````python
fig,axes=plt.subplots(1,2,figsize=(14,5))
[axes[0].plot(df[df.country==c].year,df[df.country==c].value,marker='o',ms=3,label=c) for c in cc]
axes[0].set_title("Trends by Country",fontweight='bold')
axes[0].legend(fontsize=7,ncol=2)
sns.heatmap(df.pivot_table('value','country','year',aggfunc='mean').round(1),ax=axes[1],cmap='YlOrRd',annot=True,fmt='.1f',annot_kws={'size':7})
axes[1].set_title("Heatmap",fontweight='bold')
plt.tight_layout()
plt.savefig('example_plot.png',dpi=150,bbox_inches='tight')
plt.show()
print("✅ Plots saved")
````
